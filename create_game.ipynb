```python
import gambit
from hotel import generate_strategy_permutations,generate_payoffs
import math
```

```python
n=5
s=3
```

```python
def generate_s_name(strategy):
    name=""
    for s in strategy:
        if s==0:
            name+="C"
        elif s==1:
            name+="DL"
        else:
            name+="DR"
        name+=" "
    return name
def GCD(a,b):
    """ The Euclidean Algorithm """
    a = abs(a)
    b = abs(b)
    while a:
        a, b = b%a, a
    return b
        
        
def GCD_List(list):
    """ Finds the GCD of numbers in a list.
    Input: List of numbers you want to find the GCD of
        E.g. [8, 24, 12]
    Returns: GCD of all numbers
    """
    return reduce(GCD, list)
```

```python
n_strategies=[]
for _ in range(n):
    n_strategies.append(s)
g = gambit.Game.new_table(n_strategies)
g.title = "A hotellings game"
for i in range(n):
    g.players[i].label = "Player "+str(i)
    g.players[i].strategies[0].label = "Cooperate"
    g.players[i].strategies[1].label = "Defect Left"
    g.players[i].strategies[2].label = "Defect Right"
strategies=generate_strategy_permutations(n)
for strategy in strategies:
    s=tuple(strategy)
    g[s].label=generate_s_name(strategy)
    payoffs=generate_payoffs(strategy,n)
    #print(payoffs)
    tot=math.factorial(n)
    payoffs=[p*tot for p in payoffs]
    #f=GCD_List(payoffs)
    #payoffs=[p/f for p in payoffs]
    #print(payoffs)
    for player,payoff in enumerate(payoffs):
        g[s][player]=int(payoff)
g
```

```python
f=open('gn3.efg','w+')
f.write(g.write())
f.close()
```
